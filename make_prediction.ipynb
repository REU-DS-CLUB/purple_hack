```python
from fastparquet import ParquetFile
import pandas as pd
import numpy as np
import os
import pickle 
```

```python
file_path = "Data/test_sber.parquet"
model_path = "AutogluonModels/ag-20240313_131005/models/WeightedEnsemble_L2/model.pkl"
submit_name = "WeightedEnsemble_L2" # for submit file name

pf = ParquetFile(file_path)
df = pf.to_pandas()
id = df["id"].values
# df = df[feature_final]

with open(model_path, 'rb') as file:
    loaded_automl = pickle.load(file)
```

```python
y_pred = loaded_automl.predict(df)
```

```python
# predictions = np.where(y_pred > 0.03, 1, 0)

submit = pd.DataFrame({
    "id": id,
    "target_bin": y_pred,
    # "target_prob": y_pred
})

submit_file_name = 'submit_files/submit_' + submit_name + '.csv'
submit.to_csv(submit_file_name)
```

For all models in directory

```python
file_path = "Data/test_sber.parquet"
models_dir = "AutogluonModels/ag-20240313_201752/models"

pf = ParquetFile(file_path)
df = pf.to_pandas()
id = df["id"].values
folders = [f for f in os.listdir(models_dir) if os.path.isdir(os.path.join(models_dir, f))]
# df = df[feature_final]
print(folders)

for delta in [0, 0.01, -0.01]:
    for submit_name in folders:
        model_path = os.path.join(models_dir, submit_name, "model.pkl")
        print(model_path)

        with open(model_path, 'rb') as file:
            loaded_automl = pickle.load(file)

        y_pred = loaded_automl.predict_proba(df)

        predictions = np.where(y_pred > 0.03 + delta, 1, 0)

        submit = pd.DataFrame({
            "id": id,
            "target_bin": predictions,
            "target_prob": y_pred
        })

        submit_file_name = 'submit_files/submit_' + submit_name + "_d_" + str(delta) + '.csv'
        submit.to_csv(submit_file_name)

        print(submit_name, " is done!", predictions[3])
```
